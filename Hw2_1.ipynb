```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
URL = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C"
data = pd.read_html(
    URL, match='Коефіцієнт народжуваності в регіонах України', thousands=".", decimal=",")[0]
print(data.head())
```

```python
print(data.shape)
```

```python
data.replace("—", np.nan, inplace=True)
data
```

```python
data.dtypes
```

```python
data["2014"] = data["2014"].astype("float64")
data["2019"] = data["2019"].astype("float64")
data.dtypes
```

```python
missing_percentage = (data.isnull().sum() / len(data)) * 100
missing_percentage
```

```python
ukr_row = data.iloc[-1, :].copy()
ukr_row
```

```python
data.drop(data.index[-1], inplace=True)
data
```

```python
data.iloc[:, 1:] = data.iloc[:, 1:].fillna(data.iloc[:, 1:].mean())
data
```

```python
result = data[data["2019"] > ukr_row["2019"]]
result
```

```python
result2 = data.sort_values(by="2014", ascending=False).iloc[0]
result2["Регіон"]
```

```python
plt.figure(figsize=(10, 6))
data.plot(kind="bar", x="Регіон", y="2019")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Регіон'], data['1970'], linestyle='--')
plt.xticks(rotation=90)
plt.title('Графік народжуваності у 1970 році')
```

```python
plt.figure(figsize=(10, 6))
years = data.columns[1:]
birth_row = data[data['Регіон'] == 'Дніпропетровська']
birth = birth_row[years].values.tolist()[0]
plt.pie(birth, labels=years, shadow=True, autopct="%.2f%%")
plt.title('Народжуваність у Дніпропетровській області за різні роки')
```
