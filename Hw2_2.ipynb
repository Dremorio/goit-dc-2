```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./2017_jun_final.csv')
df
```

```python
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
missing_percentage = (df.isnull().sum() / len(df)) * 100
missing_percentage
```

```python
some_list = missing_percentage[missing_percentage != 0].index.tolist()
for i in some_list[1:]:
    df.drop(i, axis=1, inplace=True)
df
```

```python
missing_percentage1 = (df.isnull().sum() / len(df)) * 100
missing_percentage1
```

```python
df.dropna(axis=0, inplace=True)
df.shape
```

```python
python_data = df[df["Мова програмування"] == "Python"]
python_data
```

```python
python_data.shape
```

```python
grouped = df.groupby("Посада")
grouped
```

```python
new_df = grouped.agg({"Зарплата на місяць": ["min", "max"]})
new_df
```

```python
def fill_avg_salary(row):
    return row.mean()


new_df["avg"] = new_df.apply(fill_avg_salary, axis=1)
new_df
```

```python
new_df["avg"].describe()
```

```python
new_df.to_csv('./new_tab.csv')
```

```python
sns.set_style("dark")
x = python_data['Загальний досвід роботи']
y = python_data['Зарплата на місяць']
hue = python_data['Досвід роботи на поточному місці']

plt.figure(figsize=(16, 10))
sns.scatterplot(x=x, y=y, hue=hue)

plt.xlabel('Загальний досвід роботи')
plt.ylabel('Зарплата на місяць')
plt.title('Залежність заробітньої плати від досвіду роботи')
plt.xticks(rotation=90)
```

```python
x = new_df.index
y = new_df['avg']

plt.figure(figsize=(10, 6))
sns.lineplot(x=x, y=y, color='red')

plt.xticks(rotation=90)
```

```python
sns.countplot(x="Загальний досвід роботи", data=python_data, color='green')
plt.xlabel('Загальний досвід роботи', labelpad=20)
plt.xticks(rotation=90)
```
